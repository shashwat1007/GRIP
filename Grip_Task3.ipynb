```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### <font color=red>Read the data file and see what should be the ideal number of clusters</font> 

```python
df = pd.read_csv (r'C:\Users\HP\Downloads\Iris.csv')

df.head(8)
```

#### <font color=red> Ideal number of clusters should be the number of species that are present.We can use `unique` function for this</font>

```python
df['Species'].unique()
```

#### <font color=red>There are 3 clusters. Our unsupervised learning should also give about 3 clusters ideally. Let us see what might be the outcome.</font>

#### First of all , we need to drop the column containing species.

```python
df2 = df.drop(['Species'], axis = 1)
df2.head()
```

#### <font color=red>Using this we can plot the data</font>

```python
sns.set(style = 'whitegrid')
plt.figure(figsize = (20, 10))
sns.pairplot(df2, plot_kws = {'color' : 'purple'});
```

#### <font color=red> Heatmap - For correlation plot</font>


```python
sns.set()
plt.figure(figsize = (5, 5))
sns.heatmap(df2.corr(), annot = True, cmap = 'ocean');
```

#### <font color=red>Finding the optimum number of clusters now for K-means Classification</font>

```python
x=df.iloc[:,[0,1,2,3]].values
from sklearn.cluster import KMeans
a=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=0)
    k.fit(x)
    a.append(k.inertia_)
```

```python
plt.plot(range(1, 11),a)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### <font color=red>Applying k-means to the dataset</font>

```python
k= KMeans(n_clusters = 3, init = 'k-means++',max_iter = 400, n_init = 400, random_state = 0)
y= k.fit_predict(x)
```

#### <font color=red>Visualising the clusters</font>

```python
plt.scatter(x[y== 0, 0], x[y == 0, 1],s = 100, c = 'purple', label = 'Setosa')
plt.scatter(x[y== 1, 0], x[y == 1, 1],s = 100, c = 'blue', label = 'Versicolour')
plt.scatter(x[y== 2, 0], x[y== 2, 1],s = 100, c = 'gray', label = 'Virginica')
plt.scatter(k.cluster_centers_[:, 0], k.cluster_centers_[:,1],s = 100, c = 'cyan', label = 'Centroids')
plt.legend()
```
